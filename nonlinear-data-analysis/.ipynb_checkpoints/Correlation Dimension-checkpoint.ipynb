## 相関次元について
参照：https://ocw.hokudai.ac.jp/wp-content/uploads/2016/01/ChaosFractal-2011-Note-09.pdf

$\boldsymbol { r } _ { i } \equiv \left( x \left( t _ { i } \right) , x \left( t _ { i } + \tau \right) , \cdots , x \left( t _ { i } + ( m - 1 ) \tau \right) \right) , \quad i = 1 , \cdots , N$

$p _ { i } = \frac { 1 } { N } \sum _ { j = 1 } ^ { N } \Theta \left( \epsilon - \left| \boldsymbol { r } _ { i } - \boldsymbol { r } _ { j } \right| \right)$

$D _ { 2 } = \lim _ { \epsilon \rightarrow 0 } \frac { \sum _ { i = 1 } ^ { N } p _ { i } ^ { 2 } } { \log \epsilon }$

$ \sum _ { i = 1 } ^ { N } p _ { i } ^ { 2 } = \frac { 1 } { N ^ { 2 } } \sum _ { i = 1 } ^ { N } \sum _ { j = 1 } ^ { N } \Theta \left( \epsilon - \left| \boldsymbol { r } _ { i } - \boldsymbol { r } _ { j } \right| \right) \equiv C ( \epsilon )$

$D _ { 2 } = \lim _ { \epsilon \rightarrow 0 } \frac { \log C ( \epsilon ) } { \log \epsilon }$

$C ( \epsilon ) = \epsilon ^ { D _ { 2 } }$

```python
N=len(df)
```

```python
#tt=[[f[2*i],f[2*i+1] ]  for i in  range(10) ]
m=2
xx=np.array([np.array([ f[m*i+j] for j in range(m)] ) for i in range(int(N/2)) ])
```

```python
print(len(xx))
print(xx[0],xx[1])
```

```python
eps=0.01
count=0
Clist=[]
#M=len(xx)

M=2000
rlist=[0.1,0.5, 1.0, 5, 10, 100,1000]

for r in rlist:
    count=0
    for i in range(M):
        #if(i%100==0):
        #print(i)
        x1=xx[i]
        for j in range(M):
            x2=xx[j]
            if(  np.sum(abs(x1-x2)) <r):
                count+=1
                
    Clist.append(count/pow(N,2))
```

```python
plt.figure()
plt.plot(np.log10(rlist), np.log10(Clist), 'bo')
plt.plot(np.log10(rlist), np.log10(rlist)*0.9-4 , label="0.9x-4.0")
plt.xlabel("log(r)")
plt.ylabel("log C(r)")
plt.legend()
plt.title("Correlation Dimension")
#plt.close()
```
