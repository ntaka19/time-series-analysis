# Dealer Model
- ディーラーモデルをpythonで再現



**1. 2 Dealers**
- 買い値が売値を超えると約定が起こり、市場価格が更新される。

```python
# -*- coding: utf-8 -*-
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def func(rand):
    c = 0.01
    if rand>0.5:
        return c
    else:
        return -c 
def main():
    #parameters
    x1=100
    x2=100
    T=1000000
    L=1
    n = 0
    
    p = np.zeros(T)
    for t in range(T):
        f1 = func(rd.random())
        #print(f1)
        f2 = func(rd.random())
        #print(f2)
        x1 = x1+f1
        x2 = x2+f2
        #print x1,x2
        if(abs(x1-x2)>L):
            #print "test"
            p[n]=(x1+x2)/2
            #print p[n]
            x1 = p[n]
            x2 = p[n]
            n+=1
    #plot
    x = np.array(range(0,n))
    pp = np.zeros(n)
    for i in range(n):
        pp[i]=p[i]
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    ax1.plot(x, pp)
    plt.title("Market Dyamics(2 Dealers)")
    plt.ylabel("Market price p(n)")
    plt.xlabel("Number of Execution (n)")
    
    pdif=np.zeros(n-1)
    xx=np.zeros(n-1)
    for i in range(0,n-1):
        pdif[i]=abs(pp[i+1]-pp[i])
        xx[i]=i/(n-1)
    print(pdif[0],pp[1],pp[0])
    fig2 = plt.figure()
    plt.title("CDF of p(n)-p(n-1)")
    plt.xlabel("p(n)-p(n-1)")
    plt.ylabel("logscale cdf")
    ax2 = fig2.add_subplot(111)
    ax2.plot(sorted(pdif,reverse=True),xx, label="data")
    ax2.plot(pdif,np.exp(-3*pdif),'r', label="exp(-3*x)")
    plt.legend()
    
    ax2.set_yscale('log')
    plt.show()
    
main()
```

**2. 4 dealers**

```python
# -*- coding: utf-8 -*-
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def func(rand):
    c = 0.01
    if rand>0.5:
        return c
    else:
        return -c 
def main():
    #parameters
    #number of dealers
    N = 4
    x = np.zeros(N)
    f = np.zeros(N)
    for i in range(N):
        x[i] = 100
    
    T=1000000
    L=1
    n = 0
    
    p = np.zeros(T)
    for t in range(T):
        f[0] = func(rd.random())
        f[1] = func(rd.random())
        f[2] = func(rd.random())
        f[3] = func(rd.random())
        x[0] = x[0]+f[0]
        x[1] = x[1]+f[1]
        x[2] = x[2]+f[2]
        x[3] = x[3]+f[3]
        
        xmax = 0
        xmin = 0
        for i in range(N):
            if(xmax < x[i] or i == 0):
                xmax = x[i]
        for i in range(N):
            if(xmin > x[i] or i == 0):
                xmin = x[i]
       
        if(abs(xmax-xmin)>L):
            #print "test"
            p[n]=(xmin+xmax)/2
            #print p[n]
            x[0] = p[n]
            x[1] = p[n]
            x[2] = p[n]
            x[3] = p[n]
            n+=1
    #plot
    x = np.array(range(0,n))
    pp = np.zeros(n)
    for i in range(n):
        pp[i]=p[i]
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    ax1.plot(x, pp)
    plt.title("Market Dyamics(4 Dealers)")
    plt.ylabel("Market price p(n)")
    plt.xlabel("Number of Execution (n)")
    
    pdif=np.zeros(n-1)
    xx=np.zeros(n-1)
    for i in range(0,n-1):
        pdif[i]=abs(pp[i+1]-pp[i])
        xx[i]=i/(n-1)
    print(pdif[0],pp[1],pp[0])
    fig2 = plt.figure()
    plt.title("CDF of p(n)-p(n-1)")
    plt.xlabel("p(n)-p(n-1)")
    plt.ylabel("logscale cdf")
    ax2 = fig2.add_subplot(111)
    ax2.plot(sorted(pdif,reverse=True),xx, label="data")
    ax2.plot(pdif,np.exp(-7*pdif),'r', label="exp(-7*x)")
    plt.legend()
    
    ax2.set_yscale('log')
    plt.show()
    
main()
```

**3. Trend-follow model**

```python
# -*- coding: utf-8 -*-
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def func(rand):
    c = 0.0001
    if rand>0.5:
        return c
    else:
        return -c 
def main():
    #parameters
    #number of dealers
    N = 2
    x = np.zeros(N)
    f = np.zeros(N)
    for i in range(N):
        x[i] = 100
        
    T=5000000
    L=0.01
    n = 0
 
    d = 1.25*0.0001
    deltaP = 0
    sum1 = 0
    
    p = np.zeros(T)
    for t in range(T):
        f[0] = func(rd.random())
        f[1] = func(rd.random())
        
        M = 1
        sum1=0
        if n<M:
            deltaP = 0
        else:
            for k in range(M-1):
                sum1 += (M-k)*(p[n-k]-p[n-k-1])
            deltaP = 2*sum1/(M*(M+1))
        
        x[0] = x[0]+d*deltaP+f[0]
        x[1] = x[1]+d*deltaP+f[1]

        if(abs(x[1]-x[0])>L):
            p[n]=(x[1]+x[0])/2
            x[0] = p[n]
            x[1] = p[n]
            n+=1
    #plot time series
    x = np.array(range(0,n))
    pp = np.zeros(n)
    for i in range(n):
        pp[i]=p[i]
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    ax1.plot(x, pp)
    plt.title("Market Dyamics(2 Dealers)")
    plt.ylabel("Market price p(n)")
    plt.xlabel("Number of Execution (n)")
    
    #plot cdf
    pdif=np.zeros(n-1)
    xx=np.zeros(n-1)
    for i in range(0,n-1):
        pdif[i]=abs(pp[i+1]-pp[i])
        xx[i]=i/(n-1)

    fig2 = plt.figure()
    plt.title("CDF of p(n)-p(n-1)")
    plt.xlabel("p(n)-p(n-1)")
    plt.ylabel("logscale cdf")
    ax2 = fig2.add_subplot(111)
    ax2.plot(sorted(pdif,reverse=True),xx, label="data")
    ax2.plot(pdif,(0.1**10)*pdif**-3,'r', label="x^-3")
    plt.legend()
    
    ax2.set_xscale('log')
    ax2.set_yscale('log')
    plt.xlim(0.0001,1)
    plt.show()
    
main()
```
