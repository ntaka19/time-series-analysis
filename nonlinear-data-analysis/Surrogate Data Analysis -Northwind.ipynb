### Amplitude adjusted Fourier transform (AAFT) surrogates

```python
df = pd.read_csv(r'northwind20041025130835.csv',header=None)
ts=np.array(df.values,dtype='float').flatten()
t=np.arange(0,len(df.values))
```

```python
plt.figure()
plt.rcParams['font.family'] = ""
plt.rcParams['font.size'] = 17
#plt.subplot(121)
plt.plot(t, ts, label='f(n)')
plt.xlabel("Time", fontsize=20)
plt.ylabel("Signal", fontsize=20)
plt.title("original data")
plt.grid()
plt.show()
```

```python
def ftsurrogate(ts):
    #ts=f
    ts_fourier  = np.fft.rfft(ts)
    random_phases = np.exp(np.random.uniform(0,np.pi, int(len(ts)/2+1))*1.0j)
    ts_fourier_new = ts_fourier*random_phases
    new_ts = np.fft.irfft(ts_fourier_new)
    return new_ts

def aaftsurrogate(data):    
    N=len(data)
    r=np.random.normal(0,1,N)
    #print(data.dtype)
    #r_sort=r[np.array(data)]
    index=np.argsort(data)
    r_sort=r[index]

    r_sort_index=np.argsort(ftsurrogate(r_sort))
    return data[r_sort_index]

aaft_ts=aaftsurrogate(ts)
```

```python
plt.figure()
plt.rcParams['font.family'] = ""
plt.rcParams['font.size'] = 17
#plt.subplot(121)

plt.plot(t, aaft_ts, label='f(n)')
plt.xlabel("Time", fontsize=20)
plt.ylabel("Signal", fontsize=20)
plt.grid()
plt.show()
```

```python
aaft=pd.DataFrame(aaft_ts)
autoco=pd.plotting.autocorrelation_plot(aaft)
```
